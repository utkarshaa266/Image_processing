```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("C:/Users/dell/Pictures/lena.jpeg")
```

```python
plt.imshow(img)
```

```python
r,g,b=cv2.split(img)
```

```python
gray_avg=(r+g+b)/3
```

```python
plt.imshow(gray_avg)
```

```python
imga=img.astype(np.float32)
```

```python
r,g,b=cv2.split(imga)
```

```python
gray_avg1=(r+g+b)/3
```

```python
plt.imshow(gray_avg1)
```

```python
gray_lum=(0.29912*r)+(0.587*g)+(0.114*b)
```

```python
plt.imshow(gray_lum)
```

```python
gray_luma=(0.2126*r)+(0.7152*g)+(0.0722*b)
```

```python
plt.imshow(gray_luma)
```

```python
gray_light= (np.minimum(np.minimum(r,g),b) + np.maximum(np.maximum(r,g),b))
```

```python
plt.imshow(gray_light)
```
